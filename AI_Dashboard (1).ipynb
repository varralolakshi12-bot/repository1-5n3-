```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("tips")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
total_bill = df['total_bill'].sum()
total_tip = df['tip'].sum()
total_people = df['size'].sum()
average_bill = df['total_bill'].mean()

print("Total Bill:", total_bill)
print("Total Tip:", total_tip)
print("Total People:", total_people)
print("Average Bill:", average_bill)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="day", y="total_bill", data=df)

plt.title("Total Bill by Day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="day", y="total_bill", data=df)

plt.title("Bill Distribution by Day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x="total_bill", y="tip", data=df)

plt.title("Tip vs Total Bill")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="day", data=df)

plt.title("Number of Bills per Day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="smoker", y="tip", data=df)

plt.title("Tip Comparison: Smoker vs Non-Smoker")
plt.show()
```

# EDA (Exploratory Data Analysis)

In this section, the dataset is loaded and basic information is checked.

- The dataset contains restaurant tips data.
- We checked data structure using head(), info(), and isnull().
- No major missing values were found in the dataset.
- The dataset is clean and ready for analysis.


# Visualizations

In this section, different visualizations are created to understand the data better.

- Bar plot shows total bill distribution across different days.
- Box plot shows variation of bills for each day.
- Scatter plot shows relationship between total bill and tip.
- Count plot shows number of bills per day.
- Comparison between smokers and non-smokers is also visualized.

These graphs help to understand patterns in customer behavior.


# Insights

From the analysis, the following insights are observed:

- Smoking status does not have a strong impact on tipping behavior.
- Total bill is the main factor influencing tip amount.
- Higher bills generally lead to higher tips.
- Customer activity is higher on weekends, especially Sunday.
- Most transactions fall in the mid-range bill category.
- Tip distribution remains fairly consistent across different groups.


# Conclusion

The analysis of the tips dataset shows that customer behavior is mainly influenced by the total bill amount and day of the week.

Smoking status does not significantly affect tipping patterns.

Overall, customers tend to give higher tips for higher bills, and weekends show more activity compared to weekdays.
